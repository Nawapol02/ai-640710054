```python
import pandas as pd
```

```python
df = pd.read_csv("student.csv")
```

```python
df
```

หาค่าที่หายไปว่ามีอยู่ในตารางไหนบ้าง

```python
df.isnull().sum()
```

แทนที่ข้อมูลหายที่เป็นตัวเลขด้วยค่าเฉลี่ยก่อนเพราะว่าสามารถจัดการข้อมูลได้ง่ายและไม่กระทบกับข้อมูลอื่น

```python
mean_age = df['อายุ'].mean()
df['อายุ'].fillna(mean_age, inplace=True)

mean_gpa = df['GPA'].mean()
df['GPA'].fillna(mean_gpa, inplace=True)

mean_height = df['ส่วนสูง_cm'].mean()
df['ส่วนสูง_cm'].fillna(mean_height, inplace=True)

mean_weight = df['น้ำหนัก_kg'].mean()
df['น้ำหนัก_kg'].fillna(mean_weight, inplace=True)

mean_hour = df['ชั่วโมงเรียน_ต่อสัปดาห์'].mean()
df['ชั่วโมงเรียน_ต่อสัปดาห์'].fillna(mean_hour, inplace=True)
```

ทำการบันทึกค่าและปัดเศษกลับไปใน dataframe ถ้าไม่ทำข้อมูลที่แก้ไขจะแสดงทศนิยมหลายตำแหน่งเกินไป

```python
df['อายุ'] = df['อายุ'].round(1)
df['GPA'] = df['GPA'].round(2)
df['ส่วนสูง_cm'] = df['ส่วนสูง_cm'].round(1)
df['น้ำหนัก_kg'] = df['น้ำหนัก_kg'].round(1)
df['ชั่วโมงเรียน_ต่อสัปดาห์'] = df['ชั่วโมงเรียน_ต่อสัปดาห์'].round(1)
```

```python
df
```

ทำการแก้ไขข้อมูลเพศและคณะ ให้เป็นชื่อมาตราฐานเพื่อจัดกลุ่มให้กับข้อมูลประเภทเดียวกัน

```python
def standardize_gender(name1):
    if name1 in ['ชาย', 'ช', 'Male', 'M']:
        return 'ชาย'
    elif name1 in['หญิง', 'ญ', 'Female', 'F']:
        return 'หญิง'
    else:
        return name1

def standardize_faculty(name2):
    if name2 in ['วิทยาศาสตร์', 'วิทย์', 'Science', 'คณะวิทยาศาสตร์', 'Sci']:
        return 'วิทยาศาสตร์'
    elif name2 in['วิศวะ', 'Engineering', 'วิศวกรรม', 'วิศวฯ', 'วิศวกรรมศาสตร์']:
        return 'วิศวกรรมศาสตร์'
    elif name2 in ['ศิลปศาสตร์', 'ศิลป์', 'Arts', 'Liberal Arts']:
        return 'ศิลปศาสตร์'
    else:
        return name2
```

```python
df['เพศ'] = df['เพศ'].apply(standardize_gender)
df['คณะ'] = df['คณะ'].apply(standardize_faculty)
```

```python
df
```

จากนั้นทำการเติมค่าที่พบบ่อยให้กับข้อมูลเพศและคณะเพื่อจัดกลุ่มข้อมูลให้ครบถ้วน

```python
df['เพศ'].fillna(df['เพศ'].mode()[0], inplace=True)
df['คณะ'].fillna(df['คณะ'].mode()[0], inplace=True)
```

```python
df
```

ลบข้อมูลที่ไม่มีชื่อออกเพื่อตัดข้อมูลที่ไม่สามารถหาค่ามาแทนได้

```python
df = df.dropna()
```

```python
df
```

ลบข้อมูลที่มีค่าผิดปกติเพื่อตัดค่าที่ออกนอกกลุ่มข้อมูลส่วนใหญ่มากเกินไป

```python
Q1 = df['อายุ'].quantile(0.25)
Q3 = df['อายุ'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_no_outliers = df[(df['อายุ'] >= lower_bound) & (df['อายุ'] <= upper_bound)]
```

```python
df_no_outliers
```

ทำการปรับลำดับของตาราง

```python
df_final = df_no_outliers.reset_index(drop=True)
```
